# Facebook DATA

```python
#Upload dataFrame

df_FB=pd.read_csv("C:/Users/ElAbed/PCD(databases)/FB.csv")
data_FB=df_FB.drop(['Date'],axis=1)
# Create standardizer
standardizer_FB = StandardScaler()

# Fit standardizer to dataframe
standardizer_FB.fit(data_FB)
# Apply to data
data_std_FB=standardizer_FB.transform(data_FB)
data_train_FB,data_test_FB = train_test_split(data_std_FB ,test_size=0.3, shuffle=False)
train_FB,test_FB=train_test_split(data_FB ,test_size=0.3, shuffle=False)

X_train_FB = []
Y_train_FB = []
for i in range(len(data_train_FB)):
    X_train_FB.append([data_train_FB[i][0],data_train_FB[i][1],data_train_FB[i][2],data_train_FB[i][5]])
    Y_train_FB.append(data_train_FB[i][3])

X_train_FB.remove(X_train_FB[len(X_train_FB)-1])
Y_train_FB.remove(Y_train_FB[0])


X_train_FB =np.array(X_train_FB) # Open,Low,High,Volume
Y_train_FB=np.array(Y_train_FB) # Close

X_train_FB[len(X_train_FB)-1]

X_test_FB = []
Y_test_FB = []
for i in range(len(data_test_FB)):
    X_test_FB.append([data_test_FB[i][0],data_test_FB[i][1],data_test_FB[i][2],data_test_FB[i][5]])
    Y_test_FB.append(data_test_FB[i][3])

X_test_FB.remove(X_test_FB[len(X_test_FB)-1])
Y_test_FB.remove(Y_test_FB[0])

X_test_FB =np.array(X_test_FB) 
Y_test_FB=np.array(Y_test_FB)

regressor_FB = LinearRegression()
regressor_FB.fit(X_train_FB,Y_train_FB)
#Linear Regression Prediction
predicted_LR_FB = regressor_FB.predict(X_test_FB)
#SVR Prediction
model_SVR.fit( X_train_FB , Y_train_FB )
predicted_SVR_FB = model_SVR.predict( X_test_FB )
#LSTM Prediction
#predicted_LSTM_FB = model_LSTM.predict(X_test_FB)
```

```python
#Facebook Data for LSTM
#Train
X_train_lstm_FB=[]### List of 60 days of Open , Low , High ,Volume
Y_train_lstm_FB=[]### List of close
for i in range (60,data_train_FB.shape[0]):
    list_lstm_FB=[]
    for j in range(60):
        list_lstm_FB.append([data_train_FB[i-60:i][j][0],data_train_FB[i-60:i][j][1],data_train_FB[i-60:i][j][2],data_train_FB[i-60:i][j][5]])
    X_train_lstm_FB.append(list_lstm_FB)
    Y_train_lstm_FB.append(data_train_FB[i,3])
X_train_lstm_FB=np.array(X_train_lstm_FB)
Y_train_lstm_FB=np.array(Y_train_lstm_FB)

#Test
X_test_lstm_FB=[]### List of 60 days of Open , Low , High ,Volume
Y_test_lstm_FB=[]### List of close

df_lstm_FB=np.concatenate((data_train_FB[1499:],data_test_FB ), axis=0)
for i in range (60,df_lstm_FB.shape[0]):
    list_lstm_FB=[]
    for j in range(60):
        list_lstm_FB.append([df_lstm_FB[i-60:i][j][0],df_lstm_FB[i-60:i][j][1],df_lstm_FB[i-60:i][j][2],df_lstm_FB[i-60:i][j][5]])
    X_test_lstm_FB.append(list_lstm_FB)
    Y_test_lstm_FB.append(df_lstm_FB[i,3])
X_test_lstm_FB=np.array(X_test_lstm_FB)
Y_test_lstm_FB=np.array(Y_test_lstm_FB)

predicted_LSTM_FB = model_LSTM.predict(X_test_lstm_FB)
```
